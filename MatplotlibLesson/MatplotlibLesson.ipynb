```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,20)
y = x * x
z = x + y
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Useless graph")
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 1")
plt.subplot(3,2,2)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 2")
plt.subplot(3,2,3)
plt.plot(x,z)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 3")
plt.subplot(3,2,4)
plt.plot(y,z)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 4")
plt.subplot(3,2,5)
plt.plot(z,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 5")
plt.subplot(3,2,6)
plt.plot(x,x)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 6")
plt.tight_layout()
```

```python
plt.plot(x,y)
plt.title("Standard graph plot")
```

```python
plt.scatter(x,y)
plt.title("Scatter plot")
```

```python
plt.hist(x,y)
plt.title("Histogram plot")
```

```python
plt.bar(x,y)
plt.title("Bar plot")
```

```python
plt.fill(x,y)
plt.title("Fill plot")
```

```python
plt.polar(x,y)
plt.title("Polar plot")
```

```python
plt.plot(x,y, label = "first example")
plt.legend(loc = 2)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1,0.1,0.8,0.8])
x1.plot(x,y)
x2 = fig.add_axes([0.1,0.1,0.3,0.3])
x2.plot(y,x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,10))
axes[0].plot(x,y)
axes[0].set_xlabel("X axis")
axes[0].set_ylabel("Y axis")
axes[0].set_title("Graph at index 0")
axes[1].plot(y,x)
axes[1].set_xlabel("X axis")
axes[1].set_ylabel("Y axis")
axes[1].set_title("Graph at index 1")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Some wierd-ass graph")
axes.plot(x,y, color = "g", lw = 1, alpha = 0.5, linestyle = "-", marker = "o", markersize = 10, markerfacecolor = "r", markeredgecolor = "g")
axes.set_xlim(0,6)
axes.set_ylim(0,60)
```

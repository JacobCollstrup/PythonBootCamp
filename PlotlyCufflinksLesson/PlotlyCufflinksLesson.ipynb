```python
import numpy as np
import pandas as pd
import plotly as pl
import cufflinks as cf
import plotly.offline as po
```

```python
po.init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.rand(100,5), columns=["a", "b", "c", "d", "e"])
```

```python
df2 = pd.DataFrame({"x":[3, 4, 7, 8, 12], "y":[1, 2, 3, 4, 5], "z": [6, 7, 8, 4, 3]})
```

```python
df3 = pd.DataFrame({"x":["a", "b", "c", "d", "e"], "y":[1, 2, 3, 4, 5], "z": [6, 7, 8, 4, 3]})
```

```python
df1.iplot(kind="scatter", x="a", y="b", mode="markers", size = 10, symbol = "x", colorscale="paired")
```

```python
df1.scatter_matrix()
```

```python
df2.iplot(kind="surface")
```

```python
cf.datagen.sinwave(10,.25).iplot(kind='surface',theme='solar',colorscale='brbg',title='Cufflinks - Surface Plot', margin=(0,0,0,0))
```

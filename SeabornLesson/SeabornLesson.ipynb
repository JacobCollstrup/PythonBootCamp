```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

Point plots, single parameter plots.

```python
sns.distplot(diamonds["price"], kde=False)
```

```python
sns.kdeplot(diamonds["price"])
```

Distribution plots:

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "hex", color = "#4CB391")
```

```python
dots = sns.load_dataset("dots")
```

```python
dots.head()
```

```python
sns.pairplot(dots, hue="choice")
```

```python
tips.head()
```

```python
x = sns.pairplot(tips, hue = "day")
x.map_diag(sns.distplot)
x.map_upper(sns.kdeplot)
```

Categorial plots:

```python
sns.stripplot(x="smoker",y="tip", data=tips, hue="sex", palette="viridis")
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex")
```

Matrix plots:

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
tips.head()
```

```python
flights2 = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
tips2 = tips.pivot_table(index="day", columns="tip", values="total_bill")
```

```python
sns.heatmap(flights2, lw=1, cmap="rocket")
```

```python
sns.clustermap(flights2, cmap="cool")
```

Grids

```python
sns.countplot(x="smoker", data = tips)
sns.set_context("notebook")
sns.despine(left=True, bottom=True, top=False)
sns.set_style("ticks")
```
